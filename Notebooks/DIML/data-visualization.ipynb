### Question 1

```python


# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
X = heart_disease.data.features 
y = heart_disease.data.targets 
  
print(X.head())
print(y.head())
```

```python


pyplt.hist(X['age'], bins=20)
pyplt.xlabel('Age')
pyplt.ylabel('Count')
pyplt.title('Distribution the Number of people with heart disease and their Age')
pyplt.show()
```

```python


sbn.boxplot(x=y.iloc[:,0], y=X['chol'])

pyplt.xlabel('Heart Disease Severity')
pyplt.ylabel('Cholesterol Level')
pyplt.title('Distribution showing Cholesterol Level by Heart Disease Severity')
pyplt.show()
```

```python
sbn.scatterplot(x=X['age'], y=X['chol'], hue=y.iloc[:,0])

pyplt.xlabel('Age')
pyplt.ylabel('Cholesterol')
pyplt.title('Distribution showing Age vs. Cholesterol by Heart Disease Severity')
pyplt.show()
```
